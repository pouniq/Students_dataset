```python
df_center = pd.read_csv('center-1.csv')
df_math = pd.read_csv('math.csv')
df_sci = pd.read_csv('sci.csv')
df_hol = pd.read_csv('holy.csv')
df_per = pd.read_csv('persian.csv')
df_social = pd.read_csv('social.csv')
df_write = pd.read_csv('writing.csv')
```

```python
df_sci = df_sci.drop(['Unnamed: 8',	'Unnamed: 9'	,'Unnamed: 10'] , axis =1)
df_sci = df_sci.drop([36,37,38,39,40,41,42])

df_center = df_center.drop('Unnamed: 0' , axis =1)
```

```python
renameM = {
    'kk': 4 , 
    'KK':4 , 
    'k': 3 ,
    'K': 3,
    'gg':2 ,
    'n':1,
    '-':0
}


datasets_dict = {
    'math': df_math,
    'sci': df_sci,
    'hol': df_hol,
    'per': df_per,
    'social': df_social,
    'write': df_write
}

for name, df in datasets_dict.items():
    df.replace(renameM, inplace=True)
    df[f'sum_{name}'] = df.sum(axis=1)
```
