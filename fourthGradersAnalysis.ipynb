```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_center = pd.read_csv('center.csv')
df_math = pd.read_csv('math.csv')
df_sci = pd.read_csv('sci.csv')
df_hol = pd.read_csv('holy.csv')
df_per = pd.read_csv('persian.csv')
df_social = pd.read_csv('social.csv')
df_write = pd.read_csv('writing.csv')
```

```python
df_center.head()
```

```python
df_center['sup'].unique()
```

```python
rename = {
    'father' : 1 ,
    'mother' : 2 ,
    'both' : 3 ,
    'other': 4
}

df_center['sup'] = df_center['sup'].replace(rename)
df_center
```

```python
df_sci = df_sci.drop(['Unnamed: 8',	'Unnamed: 9'	,'Unnamed: 10'] , axis =1)
df_sci = df_sci.drop([36,37,38,39,40,41,42])
```

```python
df_sci
```

```python


renameM = {
    'kk': 4 , 
    'KK':4 , 
    'k': 3 ,
    'K': 3,
    'gg':2 ,
    'n':1,
    '-':0
}


datasets_dict = {
    'math': df_math,
    'sci': df_sci,
    'hol': df_hol,
    'per': df_per,
    'social': df_social,
    'write': df_write
}

for name, df in datasets_dict.items():
    df.replace(renameM, inplace=True)
    df[f'sum_{name}'] = df.sum(axis=1)
```

```python
df_sci
```

```python
df_center['math'] = df_math['sum_math']
df_center['sci'] = df_sci['sum_sci']
df_center['hol'] = df_hol['sum_hol']
df_center['per'] = df_per['sum_per']
df_center['social'] = df_social['sum_social']
df_center['write'] = df_write['sum_write']
```
